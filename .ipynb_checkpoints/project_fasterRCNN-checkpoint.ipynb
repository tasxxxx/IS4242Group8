```python
# lib importation
import os
import json
import torch
import torch.optim as optim
from torchvision.models.detection import fasterrcnn_resnet50_fpn
from torchvision.transforms import functional as F
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import random
```

```python
# obtain the directory for train, validation, test
data_directory = os.path.join(os.getcwd(), 'data_preprocessing_image_classification')

train_directory = os.path.join(data_directory, 'train')
validation_directory = os.path.join(data_directory, 'validation')
test_directory = os.path.join(data_directory, 'test')
```

```python
# obtain the img, box and label for images
def obtain_image_tensor(directory):
    files = [file for file in os.listdir(directory) if file != '.DS_Store']
    label_map = {}
    index = 1
    img_index = 0
    
    image_dict = {}
    box_dict = {}
    label_dict = {}
    
    
    for file in files:
        label_map[file] = index
        new_directory = os.path.join(directory, file)
        images = os.listdir(new_directory)
        jpg_images = [image for image in images if image.endswith('.jpg')]
        for image in jpg_images:
            # convert img to tensor
            img_tensor = F.to_tensor(Image.open(os.path.join(new_directory, image)))
            img_tensor = img_tensor.unsqueeze(0)
            image_dict[img_index] = img_tensor
            
            # convert bounding boxes to tensor
            box = [0,0,127,127] #each img is each item so the size is the bounding box
            box_tensor = torch.tensor([box], dtype = torch.float32)
            box_dict[img_index] = box_tensor
            
            # convert label to tensor
            label = index 
            label_tensor = torch.tensor([label])
            label_dict[img_index] = label_tensor
            
            img_index += 1
            
        #increment index for label_map
        index += 1
    
    return label_map, image_dict, box_dict, label_dict
            
label_map, image_dict_train, box_dict_train, label_dict_train = obtain_image_tensor(train_directory)
label_map, image_dict_validation, box_dict_validation, label_dict_validation = obtain_image_tensor(validation_directory)
label_map, image_dict_test, box_dict_test, label_dict_test = obtain_image_tensor(test_directory)
```

```python
# reshuffle the data so that learning is from different classes 
random.seed(42) # Set seed for reproducibility

# Randomize the order of keys
keys = list(image_dict_train.keys())
random.shuffle(keys)

# Create a new dictionary with shuffled keys
image_dict_train_shuffled = {key: image_dict_train[key] for key in keys}
```

```python
# Define the model
model = fasterrcnn_resnet50_fpn(pretained=False) # setting pretrained to false for self_trained models
model.train() # set to train mode

optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
```

```python
num_epochs = 1
index = 0
for epoch in range(num_epochs):
    total_loss = 0
    for key, image_tensor in image_dict_train_shuffled.items():
        
        # create the target for model input formatting
        target = {'boxes': None , 'labels': None}
        
        # extract the box and label
        box = box_dict_train[key]
        label = label_dict_train[key]
        
        target['boxes'] = box
        target['labels'] = label
        targets = [target]
        
        #set grad = 0
        optimizer.zero_grad()
        
        #forward pass in model
        loss_dict = model(image_tensor, targets)
    
        losses = sum(loss for loss in loss_dict.values())
        
        # Backward pass
        losses.backward()
        
        optimizer.step()
        total_loss += losses.item()
        print(f"Training: {index} / {len(image_dict_train_shuffled)} completed")
        index += 1
        
    print(f"Epoch {epoch+1}, Avg Loss: {total_loss / len(image_dict_train_shuffled)}")
```

```python
# Save the entire model
torch.save(model, 'model_RCNN.pth')
```

```python
# Load the entire model
model = torch.load('model_RCNN.pth')
```

```python
# set model to evaluation mode
model.eval()

prediction_result = {}
index = 0

for key, image_tensor in image_dict_test.items():
    with torch.no_grad():
        predictions = model(image_tensor)
        
        prediction_model = predictions[0]
        top_prediction_box = prediction_model['boxes'][0].tolist() # obtain top box coordinates
        
        prediction_labels = prediction_model['labels'].tolist() # obtain top label 
        top_prediction_label = prediction_labels[0]
        
        prediction_scores = prediction_model['scores'].tolist() # obtain top score
        top_prediction_score = prediction_scores[0]
        
        prediction_result[index] = {'box': top_prediction_box, 'label': top_prediction_label, 'score': top_prediction_score}
        index += 1
```

```python
reversed_label_map = {}

for key, value in label_map.items():
    reversed_label_map[value] = key

reversed_label_map
```

```python
true_labels = []
predicted_labels = []

for key, value in label_dict_test.items():
    actual_label = value.tolist()
    actual_label = reversed_label_map[actual_label[0]]
    true_labels.append(actual_label)
    
for key, value in prediction_result.items():
    prediction_label = value['label']
    prediction_label = reversed_label_map[prediction_label]
    predicted_labels.append(prediction_label)
    
class_labels = ['glass','metal','paper','plastic']
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
import seaborn as sns

# Generate the classification report
report = classification_report(true_labels, predicted_labels, zero_division=0)

# Print the classification report
print("Classification Report:")
print(report)

# Generate confusion matrix
cm = confusion_matrix(true_labels, predicted_labels, labels=class_labels)

# Calculate accuracy
accuracy = accuracy_score(true_labels, predicted_labels)
print("\nAccuracy:", accuracy)

# Visualize confusion matrix
plt.figure(figsize=(12, 8))
heatmap = sns.heatmap(cm, annot=True, fmt='d', cmap='crest', xticklabels=class_labels, yticklabels=class_labels)
heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, fontsize=12)
heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, fontsize=12)
plt.title('Confusion Matrix', fontsize=15, pad=12, color='darkblue')
plt.ylabel('True label', fontsize=14)
plt.xlabel('Predicted label', fontsize=14)
plt.show()
```

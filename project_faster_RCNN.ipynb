```python
# lib importation
import os
import json
import torch
import torch.optim as optim
from torchvision.models.detection import fasterrcnn_resnet50_fpn
from torchvision.transforms import functional as F
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
```

```python
# obtain the directory for train, validation, test
data_directory = os.path.join(os.getcwd(), 'data_preprocessing_image_classification')

train_directory = os.path.join(data_directory, 'train')
validation_directory = os.path.join(data_directory, 'validation')
test_directory = os.path.join(data_directory, 'test')
```

```python
# obtain the img, box and label for images
def obtain_image_tensor(directory):
    files = os.listdir(directory)
    label_map = {}
    index = 1
    img_index = 0
    
    image_dict = {}
    box_dict = {}
    label_dict = {}
    
    
    for file in files:
        label_map[file] = index
        new_directory = os.path.join(directory, file)
        images = os.listdir(new_directory)
        jpg_images = [image for image in images if image.endswith('.jpg')]
        for image in jpg_images:
            # convert img to tensor
            img_tensor = F.to_tensor(Image.open(os.path.join(new_directory, image)))
            img_tensor = img_tensor.unsqueeze(0)
            image_dict[img_index] = img_tensor
            
            # convert bounding boxes to tensor
            box = [0,0,127,127] #each img is each item so the size is the bounding box
            box_tensor = torch.tensor(box, dtype = torch.float32)
            box_dict[img_index] = box_tensor
            
            # convert label to tensor
            label = index 
            label_tensor = torch.tensor(label)
            label_dict[img_index] = label_tensor
            
            img_index += 1
            
        #increment index for label_map
        index += 1
    
    return label_map, image_dict, box_dict, label_dict
            
label_map, image_dict, box_dict, label_dict = obtain_image_tensor(train_directory)
```

```python
# obtaining the values for processed train images
with open (processed_train_json_path, 'r') as file:
    data = json.load(file)
    
    image_paths = {}
    boxes = {}
    labels = {}
    label_map = {}
    
    # to obtain the label mapping
    for category in data['categories']:
        label_id = category['id']
        label_category = category['name']
        label_map[label_id] = label_category
    
    # to obtain the image file name
    for image in data['images']:
        image_id = image['id']
        image_name = image['file_name']
        image_paths[image_id] = image['file_name']
    
    # to obtain the bounding boxes and ground truth
    # 'bbox': [x_min, y_min, width, height], 'category_id' : number (to represent the class) 
    for annotation in data['annotations']:
        image_id = annotation['image_id']
        
        coordinates = annotation['bbox'] # obtain the bbox value
        x_min = coordinates[0]
        y_min = coordinates[1]
        x_max = x_min + coordinates[2]
        y_max = y_min + coordinates[3]
        
        label = annotation['category_id']
        
        # check if image_id alr exist (meaning the image_id has multiple boxes)
        if image_id in boxes:
            boxes[image_id].append([x_min, y_min, x_max, y_max])
        else:
            boxes[image_id] = [[x_min, y_min, x_max, y_max]]
            
        if image_id in labels:
            labels[image_id].append(label)
        else:
            labels[image_id] = [label]
```

```python
num_epochs = 1 #each epoch ard 40min
for epoch in range(num_epochs):
    total_loss = 0
    for img_id in image_paths:
        
        #create the target for model input formatting
        targets = [{'boxes': None , 'labels': None}]
        
        #convert img to tensor for model
        image_name = image_paths[img_id]
        img_tensor = F.to_tensor(Image.open(os.path.join(processed_train_path, image_name)))
        img_tensor = img_tensor.unsqueeze(0)
        
        #convert boxes to tensor for model
        box_list = boxes[img_id]
        box_list_tensor = torch.tensor(box_list, dtype = torch.float32)

        #convert labels to tensor for model
        label_list = labels[img_id]
        label_list_tensor = torch.tensor(label_list)
        
        #input the tensor into targets
        targets[0]['boxes'] = box_list_tensor
        targets[0]['labels'] = label_list_tensor
        
        #set grad = 0
        optimizer.zero_grad()
        
        #forward pass in model
        loss_dict = model(img_tensor, targets)
    
        losses = sum(loss for loss in loss_dict.values())
        
        # Backward pass
        losses.backward()
        
        optimizer.step()
        total_loss += losses.item()
        print(f"Training: {img_id} / {len(image_paths)} completed")
        
    print(f"Epoch {epoch+1}, Avg Loss: {total_loss / len(image_paths)}")
```
